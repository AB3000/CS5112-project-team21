```python
import matplotlib.pyplot as plt
```

```python
def plotfn(time, text_len, alg_name, title):
    for k in range(0, len(time)):
        plt.plot(text_len, time[k])
    plt.legend(alg_name)
    plt.ylabel("Time(s)")
    plt.xlabel("Text Length(#character)")
    plt.title(title)
    plt.savefig(title+".png")
```

```python
#Test1
time1 = [[1,2,3,4,5], [0.5, 1.5, 2.5, 3.5, 4]]
text_len1 = ["100","1000","10000","100000","1000000"]
alg_name1 = ["alg1", "alg2"]
title1 = "Test1"
plotfn(time1, text_len1,alg_name1, title1)
```

```python
#Test2
time2 = [[0.1,0.15,0.56,1,1.7], [0.02, 0.13, 0.8, 1.3, 2.2], [0.45, 0.56, 0.67, 0.89, 1.0]]
text_len2 = ["100","1000","10000","100000","1000000"]
alg_name2 = ["alg1", "alg2", "alg3"]
title2 = "Test2"
plotfn(time2, text_len2,alg_name2,title2)
```
